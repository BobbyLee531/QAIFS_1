```python
from networkx.algorithms.community import greedy_modularity_communities

returns = data.pct_change().dropna()
correlation_matrix = returns.corr()

G = nx.Graph()

for stock in correlation_matrix.columns:
    G.add_node(stock)

threshold = 0.65
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        correlation = correlation_matrix.iloc[i, j]
        if correlation >= threshold:
            G.add_edge(correlation_matrix.columns[i], correlation_matrix.columns[j], weight=correlation)

communities = list(greedy_modularity_communities(G))

pos = {}
y_offset = 1

#draw community by community
for i, community in enumerate(communities):
    subG = G.subgraph(community)
    sub_pos = nx.spring_layout(subG, seed=42, k=15, iterations=100)
    x_offset = i * 3

    for node, (x, y) in sub_pos.items():
        pos[node] = (x + x_offset, y + y_offset)

#deal with the isolated nodes
isolated_nodes = [node for node in G.nodes if G.degree(node) == 0]
if isolated_nodes:
    isolated_pos = {node: (i * 1, -2) for i, node in enumerate(isolated_nodes)}
    pos.update(isolated_pos)



edge_traces = []
edge_labels = []

for edge in G.edges(data=True):
    stock1, stock2, attributes = edge
    x0, y0 = pos[stock1]
    x1, y1 = pos[stock2]

    edge_traces.append(go.Scatter(
        x=[x0, x1, None], y=[y0, y1, None],
        line=dict(width=1, color='gray'),
        mode='lines',
        hoverinfo='none'
    ))

    edge_labels.append(go.Scatter(
        x=[(x0 + x1) / 2], y=[(y0 + y1) / 2],
        mode='text',
        text=[f'{attributes["weight"]:.2f}'],
        showlegend=False,
        hoverinfo='name',
        name=f'{stock1}-{stock2}:{attributes["weight"]:.2f}'

    ))

node_traces = []
for i, community in enumerate(communities):
    for node in community:
        x, y = pos[node]
        node_traces.append(go.Scatter(
            x=[x], y=[y],
            mode='markers+text',
            text=node,
            textposition="top center",
            hoverinfo='none',
            marker=dict(
                size=10,
                color=plt.cm.rainbow(i / len(communities)),
                line=dict(width=2, color='black')
            )
        ))

fig = go.Figure(data=edge_traces + edge_labels + node_traces)
fig.update_layout(
    title='Stock Correlation Network (Greedy Modularity Community Detection)',
    showlegend=False,
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    plot_bgcolor='white'
)

fig.show()
```
